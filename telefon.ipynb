```python
def getAndParseURL(url):
    result = requests.get(url, headers={"User-Agent":"Chrome/115.0.5790.171"}) # Safari/537.36. Chrome/103.0.0.0
    soup = bts(result.text, "html.parser")
    return soup
```

```python
features = []
for telefon in telefonlar:
    html = getAndParseURL(telefon)
    try:
        marka = html.find("a",{"class":"product-brand-name-with-link"}).text.strip()
    except:
        marka = np.nan
    try:
        fiyat =html.find("div",{"class":"pr-bx-nm with-org-prc"}).span.text.replace("TL","").replace(".","").strip()
    except:
        fiyat = np.nan
    try:
        kamera_cozunurluk =html.find("ul",{"class":"detail-attr-container"}).find(text=re.compile("Kamera Çözünürlüğü")).findNext().text.replace("ve","").replace("üstü","").replace("MP","").strip()
    except:
        kamera_cozunurluk = np.nan
    try:
        dahili_hafıza = html.find("ul",{"class":"detail-attr-container"}).find(text=re.compile("Dahili Hafıza")).findNext().text.replace("GB","").strip()
    except:
        dahili_hafıza  = np.nan
    try:
        ram =html.find("ul",{"class":"detail-attr-container"}).find(text=re.compile("RAM Kapasitesi")).findNext().text.replace("GB","").strip()
    except:
        ram = np.nan
    try:
        ekran_boyutu = html.find("ul",{"class":"detail-attr-container"}).find(text=re.compile("Ekran Boyutu")).findNext().text.replace("inç","").replace(",",".").strip()
    except:
        ekran_boyutu = np.nan
    try:
        pil_gucu = html.find("ul",{"class":"detail-attr-container"}).find(text=re.compile("Pil Gücü")).findNext().text.replace("ve","").replace("üstü","").strip()
    except:
        pil_gucu = np.nan
    try:
        mobil_baglanti_hizi=html.find("ul",{"class":"detail-attr-container"}).find(text=re.compile("Mobil Bağlantı Hızı")).findNext().text.replace("G","").strip()
    except:
        mobil_baglanti_hizi= np.nan
    try:
        model = html.find("ul",{"class":"detail-attr-container"}).find(text=re.compile("Cep Telefonu Modeli")).findNext().text
    except:
        model  = np.nan
    try:
         on_kamera_sayisi=html.find("ul",{"class":"detail-attr-container"}).find(text=re.compile("Ön Kamera Sayısı")).findNext().text.strip()
    except:
         on_kamera_sayisi= np.nan
    try:
         on_kamera_cozunurluk_araligi=html.find("ul",{"class":"detail-attr-container"}).find(text=re.compile("Ön Kamera Çözünürlük Aralığı")).findNext().text.replace("MP","").strip()
    except:
         on_kamera_cozunurluk_araligi= np.nan    
    try:
         arka_kamera_sayisi=html.find("ul",{"class":"detail-attr-container"}).find(text=re.compile("Arka Kamera Sayısı")).findNext().text.strip()
    except:
         arka_kamera_sayisi= np.nan         
    try:
         ekran_boyutu_araligi=html.find("ul",{"class":"detail-attr-container"}).find(text=re.compile("Ekran Boyut Aralığı")).findNext().text.replace("''","").strip()
    except:
         ekran_boyutu_araligi= np.nan
    try:
         isletim_sistemi= html.find("ul",{"class":"detail-attr-container"}).find(text=re.compile("İşletim Sistem")).findNext().text.strip()
    except:
         isletim_sistemi = np.nan    
    try:
         cift_hat = html.find("ul",{"class":"detail-attr-container"}).find(text=re.compile("Çift Hat")).findNext().text.strip()
    except:
         cift_hat = np.nan
    try:
        garanti_suresi=html.find("ul",{"class":"detail-attr-container"}).find(text=re.compile("Garanti Süresi")).findNext().text.replace("Yıl","").strip()
    except:
        garanti_suresi= np.nan
    try:
         sarj_girisi= html.find("ul",{"class":"detail-attr-container"}).find(text=re.compile("Şarj Girişi")).findNext().text.strip()
    except:
         sarj_girisi= np.nan    
    try:
         ana_kamera_cozunurluk=html.find("ul",{"class":"detail-attr-container"}).find(text=re.compile("Ön Kamera Çözünürlük Aralığı")).findNext().text.replace("MP","").strip()
    except:
         ana_kamera_cozunurluk= np.nan    
    try:
         nfc= html.find("ul",{"class":"detail-attr-container"}).find(text=re.compile("NFC")).findNext().text.strip()
    except:
         nfc= np.nan    
    try:
        kulaklık_girisi = html.find("ul",{"class":"detail-attr-container"}).find(text=re.compile("Kulaklık Giriş")).findNext().text.strip()
    except:
        kulaklık_girisi= np.nan                 
  
    try:
         batarya= html.find("ul",{"class":"detail-attr-container"}).find(text=re.compile("Batarya Kapasitesi Aralığı")).findNext().text.replace("mAh","").strip()
    except:
         batarya= np.nan    
    try:
         renk=html.find("ul",{"class":"detail-attr-container"}).find(text=re.compile("Renk")).findNext().text.strip()
    except:
         renk= np.nan    
  
            
    features.append([marka, fiyat, kamera_cozunurluk, dahili_hafıza, ram, ekran_boyutu,pil_gucu,mobil_baglanti_hizi,model,on_kamera_sayisi,on_kamera_cozunurluk_araligi,arka_kamera_sayisi,ekran_boyutu_araligi,isletim_sistemi,cift_hat,garanti_suresi,sarj_girisi,ana_kamera_cozunurluk,nfc, kulaklık_girisi,batarya,renk])
    time.sleep(1)
    
df_columns = ["marka","fiyat","kamera çözünürlük","Dahili hafıza","Ram","ekran boyutu","pil gücü","mobil bağlantı hızı","model","ön kamera sayısı","on kamera cozunurluk araligi","arka kamera sayisi","ekran boyutu aralığı","işletim sistemi","çift hat","garanti süresi","şarj girişi","ana kamera çözünürlük","nfc","kulaklık girişi","batarya","renk"]

df = pd.DataFrame.from_records(features, columns=df_columns)
```
